# Business questions

### 1.What percentage of customers are churning?

```python
# count churn values
churn_counts = df["Churn"].value_counts()
churn_counts
```

```python
#Calculate the churn rate
churn_rate = churn_counts["Yes"] / churn_counts.sum() * 100
churn_rate
```

##### Approximately 26-27% of customoers have churned, indicating a significant customer retention challenge

### 2.Are customers on different contract types more likely to churn?

```python
# Churn by Contract type(Count)
contract_churn = pd.crosstab(df["Contract"], df["Churn"])
contract_churn
```

```python
# CHurn Rate by Contract
contract_Churn_rate = (contract_churn.div(contract_churn.sum(axis=1), axis=0) * 100)
contract_Churn_rate
```

##### Insight:
##### Customers on month-to-month contracts have the highest churn rate,
##### while customers on long-term contracts (one-year and two-year) churn much less.


### 3.Do customers with higher monthly charges churn more?

```python
# Compare average monthly charges by Churn 
monthly_charge_churn = df.groupby("Churn")["MonthlyCharges"].mean()
monthly_charge_churn
```

```python
sns.boxplot(x ="Churn", y="MonthlyCharges", data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

##### Insight:
##### Customers who churn tend to have higher average monthly charges,
##### suggesting that pricing may influence customer retention.


### 4.Are newer customers more likely to churn than long-term customers?

```python
# Compare average tenure by churn
tenure_churn = df.groupby("Churn")["tenure"].mean()
tenure_churn
```

```python
sns.boxplot(x="Churn", y="tenure", data=df)
plt.title("Tenure vs Churn")
plt.show()
```

##### Insight:
##### Customers with shorter tenure are more likely to churn,
##### indicating that early-stage customer retention is critical.


### 5.Which customer segments show higher churn risk?

```python
# Churn by Internet Service
internet_churn = pd.crosstab(df["InternetService"], df["Churn"])
internet_churn
```

```python
#Convert to churn rate
internet_churn_rate = (internet_churn.div(internet_churn.sum(axis=1), axis=0) * 100)
internet_churn_rate
```

##### Insight:
##### Customers using fiber optic internet service show a higher churn rate
##### compared to other internet service types, indicating potential service or pricing issues.
